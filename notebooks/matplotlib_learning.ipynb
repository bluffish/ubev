```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x1 = np.linspace(0, 5, 10)
y1 = x1 ** 2

plt.plot(x1, y1)
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'r')
plt.subplot(1, 2, 2)
plt.plot(x1, y1, 'b')
```

```python
fig1 = plt.figure(figsize=(5, 4), dpi=100)
ax1 = fig1.add_axes([.1, .1, .9, .9])
ax1.set_xlabel('Days')
ax1.set_ylabel('Days Squared')
ax1.set_title('DSQTCH')
ax1.plot(x1, y1, label='x/x^2')
ax1.plot(y1, x1, label='x^2/x')
ax1.legend(loc=0)

ax2 = fig1.add_axes([.45, .45, 0.4, 0.3])
ax2.set_xlabel('Days')
ax2.set_ylabel('Days Squared')
ax2.set_title('DSQTCH')
ax2.plot(y1, x1, 'r', label='x^2/x')

ax2.text(0, 8, 'Message')
```

```python
fig2, axs2 = plt.subplots(1, 3, figsize=(8, 4))
plt.tight_layout()

axs2[1].set_title("Plot 2")
axs2[1].set_xlabel("X")
axs2[1].set_ylabel("Y")
axs2[1].plot(x1, y1)
```

```python
fig3 = plt.figure(figsize=(6, 4))
ax3 = fig3.add_axes([0, 0, 1, 1])
ax3.plot(x1, y1, color='navy', alpha=.75, lw=5, ls='-', marker='o', markersize=20, markerfacecolor='y', markeredgecolor='r')
ax3.set_xlim([0, 3])
ax3.set_ylim([0, 8])
ax3.grid(True, color='.6', dashes=(5, 2, 1, 2)) 
ax3.set_facecolor('#FAEBD7')
```

```python
fig3.savefig("3rdplot.png", bbox_inches='tight')
```

```python

# Labels
labels = ['AC', 'AU', 'IC', 'IU']

# Values
UCE = [294355, 10588, 9779, 5150]
FOCAL = [265120, 44630, 2096, 8026]

# Bar width
bar_width = 0.35

# Bar positions
r1 = np.arange(len(UCE))
r2 = [x + bar_width for x in r1]

# Create the bars
plt.bar(r1, UCE, width=bar_width, color='blue', alpha=.75, label='UCE')
plt.bar(r2, FOCAL, width=bar_width, color='red', alpha=.75, label='FOCAL')

# Label the bars
plt.xlabel('Frequency')
plt.xticks([r + bar_width for r in range(len(UCE))], labels)
plt.ylabel('Pixel')
plt.title('Comparison of UCE and FOCAL')

# Create the legend
plt.legend()

# Display the bar chart
plt.tight_layout()
plt.show()
```
